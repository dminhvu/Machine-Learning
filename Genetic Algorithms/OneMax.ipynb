```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
class OneMax():
    def __init__(self,individual_size,population_size,crossover_rate,elitism,n_generations,mutation_rate):
        self.individual_size = individual_size
        self.population_size = population_size
        self.crossover_rate = crossover_rate
        self.elitism = elitism
        self.n_generations = n_generations
        self.mutation_rate = mutation_rate
        self.population = []
        self.fitnesses = []
        
    def generate_population(self):
        for i in range(self.population_size):
            self.population.append(np.random.randint(0,2,self.individual_size))
    
    def compute_fitness(self,individual):
        return np.sum(individual)
    
    def sort(self):
        self.population = sorted(self.population,key=self.compute_fitness)
        
    def crossover(self,individual1,individual2):
        individual1_new = individual1.copy()
        individual2_new = individual2.copy()
        
        for i in range(self.individual_size):
            if np.random.random() < self.crossover_rate:
                individual1_new[i], individual2_new[i] = individual2[i], individual1[i]
        
        return individual1_new,individual2_new
    
    def selection(self):
        id1 = np.random.randint(0,len(self.population)-1)
        id2 = np.random.randint(0,len(self.population)-1)
        while id1 == id2:
            id2 = np.random.randint(0,len(self.population)-1)
            
        individual = self.population[id1]
        if id2 > id1:
            individual = self.population[id2]
        
        return individual
    
    def mutate(self,individual):
        for i in range(self.individual_size):
            if np.random.random() < self.mutation_rate:
                individual[i] = np.random.randint(0,2)
                
        return individual
    
    def process(self):
        self.generate_population()
        
        for i in range(n_generations):
            self.sort()
            print(f'Best: {self.compute_fitness(self.population[-1])}')
            print(f'{self.population[-1]} {self.compute_fitness(self.population[-1])}')
            self.fitnesses.append(self.compute_fitness(self.population[-1]))
            self.population = self.population[-elitism:]
            while len(self.population) < self.population_size:
                individual1 = self.selection()
                individual2 = self.selection()
                
                individual1,individual2 = self.crossover(individual1,individual2)
                
                individual1 = self.mutate(individual1)
                individual2 = self.mutate(individual2)
                self.population.append(individual1)
                if len(self.population) < self.population_size:
                    self.population.append(individual2)
        
        plt.plot(self.fitnesses)
        plt.xlabel('Generations')
        plt.ylabel('Fitness')
        plt.show()
                    
```

```python
individual_size = 30
population_size = 30
crossover_rate = 0.3
n_generations = 50
elitism = 8
mutation_rate = 0.05

onemax = OneMax(individual_size=individual_size,population_size=population_size,crossover_rate=crossover_rate,
                elitism=elitism,n_generations=n_generations,mutation_rate=mutation_rate)
onemax.process()
```
