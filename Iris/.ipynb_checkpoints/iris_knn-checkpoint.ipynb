```python
from sklearn import neighbors, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=42)
X_train[:5], X_train.shape
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
```

```python
plt.scatter(X_train[:,2],X_train[:,3],c=y_train)
plt.xlabel("petal_length")
plt.ylabel("petal_width")
```

```python
def my_weight(distances):
    # softmax
    return np.exp(distances)/np.sum(np.exp(distances))
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=12,p=5,weights=my_weight)
model.fit(X_train, y_train)
f"Accuracy: {round(model.score(X_test, y_test)*100,2)}%"
```
