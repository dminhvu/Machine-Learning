```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)
```

```python
x = np.arange(-7,7,0.01)
params = [(0, 2), (2, 1), (1, 3)]
for mu, sigma in params:
    plt.plot(normal(x, mu, sigma))
```
