```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
class MyLinearRegression:
    import numpy as np
    import sys
    from time import sleep
    
    def calculate_loss(self,y_hat,y):
        return ((y_hat-y)**2).mean()/2
    
    def calculate_gradient(self,X,y_hat,y):
        return X.T.dot(y_hat-y)
    
    def predict(self,X,w):
        return X.dot(w)
    
    def fit(self,X,y,lr=0.00001,n_epochs=100,batch_size=32):
        m = X.shape[0]
        ids = self.np.random.permutation(m)
        X_s = X[ids]
        y_s = y[ids]
        
        weights = self.np.random.rand(X.shape[1],1)
        losses = []
        
        for epoch in range(n_epochs):
            for i in range(0,m,batch_size):
                Xi = X_s[i:i+batch_size]
                yi = y_s[i:i+batch_size]
                
                y_hat = self.predict(Xi,weights)
                loss = self.calculate_loss(y_hat,yi)
                grad = self.calculate_gradient(Xi,y_hat,yi)
                
                weights -= lr*grad
                losses.append(loss)
        
        self.losses = losses
        self.weights = weights
```

# d1.csv

```python
d1 = pd.read_csv('d1.csv').values
d1.shape
```

```python
x = d1[:,:-1]
y = d1[:,-1:]
plt.plot(x,y)
```

```python
n_samples, n_features = x.shape
n_samples, n_features
```

```python
X = np.concatenate([x**2,x,np.ones((x.shape[0],1))],axis=1)
X[:5], y[:5]
```

```python
model = MyLinearRegression()
model.fit(X,y,lr=0.00001,n_epochs=100,batch_size=16)
```

```python
plt.plot(model.losses)
model.losses[-1]
```

```python
y_pred = model.predict(X,model.weights)
y_pred[:5]
```

```python
plt.plot(y,label='y')
plt.plot(y_pred,label='y_pred')
plt.legend()
```
