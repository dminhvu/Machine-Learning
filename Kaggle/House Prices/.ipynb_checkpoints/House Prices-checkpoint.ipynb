```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head(10)
```

```python
df.columns
```

```python
cols_to_delete = ['Id', 'MoSold', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'Utilities', 'Condition2', ]
df.drop(cols_to_delete,axis='columns',inplace=True)
df.head(10)
```

```python
# Fillna with max value
def fillna_with_maxval(col):
    global df
    df[col].fillna(df[col].value_counts().idxmax(),inplace=True)
```

```python
def encode_dummies(col):
    global df
    df[col] = df[col].apply(lambda x: x + col)
    dummies = pd.get_dummies(df[col])
    df = pd.concat([df, dummies],axis='columns')
    df.drop([col],axis='columns',inplace=True)
```

```python
def fillnacol(col):
    global df
    df[col].fillna(df[col].mean(),inplace=True)
```

```python
cols_to_encode = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 
                  'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
                  'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                  'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 
                  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition',
                 ]
cols_to_fillna = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt', 
                 ]
cols_to_fillna_with_maxval = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 
                              'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'MSZoning',
                              'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'SaleType', 'SaleCondition', 
                             ]
```

```python
for col in cols_to_fillna_with_maxval:
    fillna_with_maxval(col)
for col in cols_to_encode:
    encode_dummies(col)
for col in cols_to_fillna:
    fillnacol(col)
```
